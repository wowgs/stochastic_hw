```python
import numpy as np
import pandas as pd
from PIL import Image, ImageDraw
import random
```

```python
colourImg = Image.open("peppers.jpg")
colourPixels = colourImg.convert("RGB")
colourArray = np.array(colourPixels.getdata()).reshape(colourImg.size + (3,))
indicesArray = np.moveaxis(np.indices(colourImg.size), 0, 2)
allArray = np.dstack((indicesArray, colourArray)).reshape((-1, 5))


df = pd.DataFrame(allArray, columns=["y", "x", "r","g","b"])
df['l'] = 0
```

```python
class my_point():
    def __init__(self, x):
        self.r = x[0]
        self.g = x[1]
        self.b = x[2]
    def __eq__(self, other):
        return (self.r == other.r and
                self.g == other.g and
                self.b == other.b)

def dist(a, b):
    r = (a.r - b.r) ** 2
    g = (a.g - b.g) ** 2
    b = (a.b - b.b) ** 2
    return (r + g + b) ** 0.5

def new_centr(data):
    r, g, b = 0, 0, 0
    l = len(data)
    for i, x in data.iterrows():
        r += x.r
        g += x.g
        b += x.b
    r, g, b = r / l, g / l, b / l
    return my_point([r, g, b])
```

```python
def kmeans(data, k, n_iter):
    centrs = data.sample(n = k)
    centrs = [my_point([x.r, x.g, x.b]) for i, x in centrs.iterrows()]
    flag = True
    for i in range(n_iter):
        print(i)
        if not flag:
            break
        flag = False
        
        for j, x in data.iterrows():
            cur_l = x.l
            cur_d = dist(centrs[x.l], x)
            for w in range(k):
                tmp = dist(centrs[w], x)
                if tmp < cur_d:
                    x.l = w
                    cur_d = tmp
                    
        for w in range(k):
            new = new_centr(data[data.l == w])
            if centrs[w] != new:
                flag = True
                centrs[w] = new
    
    return centrs
                
    
```

```python
centrs = kmeans(df, 5, 100)
```
