```python
import numpy as np
import pandas as pd
from PIL import Image, ImageDraw
import random
import matplotlib.pyplot as plt
from sklearn.metrics import davies_bouldin_score as dbs, calinski_harabaz_score as chs
from scipy.spatial.distance import cdist, sqeuclidean, euclidean
```

```python
def kmeans(data, k, n_iter):
    centrs = np.random.choice(np.arange(len(data)), k, False)
    centrs = data[centrs]
    flag = True

        
    for i in range(n_iter):
        if not flag:
            break
        flag = False
        
        dist_ar = cdist(centrs, data, 'sqeuclidean')
        clusts = np.argmin(dist_ar, axis=0)        
                    
        for w in range(k):
            tmp = data[clusts == w]
            if len(tmp) == 0:
                new = np.array([0] * len(data[0]))
            else:    
                new = np.mean(tmp, axis=0)
            if not np.array_equal(centrs[w], new):
                flag = True
                centrs[w] = new
    
    return centrs, clusts               
```

```python
def sigma(centr, points):
    tmp = sum([sqeuclidean(centr, x) for x in points])
    tmp /= len(points)
    tmp = tmp ** 0.5
    return tmp


def db_index(data, clusts, k):
    centrs = [np.mean(data[clusts == x], axis=0) for x in range(k)]
    sigmas = [sigma(centrs[x], data[clusts == x]) for x in range(k)]
    
    ar = []
    for x in range(k):
        ar.append([])
        for y in range(k):
            if y == x:
                ar[x].append(0)
            else:
                ar[x].append((sigmas[x] + sigmas[y]) / euclidean(centrs[x], centrs[y]))
    
    ar = [max(ar[x]) for x in range(k)]
    
    return sum(ar) / k

def ch_tool(centr, data):
    return sum([np.dot(x - centr, (x - centr).T) for x in data])

def ch_index(data, clusts, k):
    glob_mean = np.mean(data, axis=0)
    centrs = [np.mean(data[clusts == x], axis=0) for x in range(k)]
    sw = sum([ch_tool(centrs[x], data[clusts == x]) for x in range(k)])
    sb = sum([np.dot(centrs[x] - glob_mean, (centrs[x] - glob_mean).T) * len(data[clusts == x]) for x in range(k)])
    return (sb / (k - 1)) / (sw / (len(data) - k))
```

```python
import os
dirs = ['res/']
for x in dirs:
    if not os.path.exists(x):
        os.makedirs(x)

max_k = 20
        
for x in ['policemen.jpg']:
    name = 'data/' + x
    image = np.array(Image.open(name))
    data = image.reshape((image.shape[0] * image.shape[1], image.shape[2]))
    res = []
    for k in range(2, max_k):
        best_centrs, best_clusts = [], []
        best_func = 10**10
        for i in range(3):
            centrs, clusts = kmeans(data, k, 200)
            func = cdist(centrs, centrs, 'sqeuclidean').max()
            if func < best_func:
                best_func = func
                best_centrs, best_clusts = centrs, clusts
        res.append([db_index(data, best_clusts, k), ch_index(data, best_clusts, k)])#, chs(data, best_clusts)])
        new_data = best_centrs[best_clusts]
        new_data.shape = image.shape
        Image.fromarray(new_data).save('res/' + str(k) + '.jpg')
```

```python
res = np.array(res)
x = list(range(2, max_k))
plt.plot(x, res[:, 0])
plt.savefig('db_index.png', dpi = 600)
plt.show()
plt.plot(x, res[:, 1])
plt.savefig('ch_index.png', dpi = 600)
plt.show()
```
