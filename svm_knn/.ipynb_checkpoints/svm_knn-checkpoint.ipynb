```python
import numpy as np
import pandas as pd
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier as knn
from sklearn.model_selection import GridSearchCV
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('chips.txt', delimiter=',', header=None, names=['a', 'b', 'y'])
data = data.sample(frac=1).reset_index(drop=True)
x = data[['a', 'b']]
y = data.y


h = 0.01
x_min, x_max = x.a.min(), x.a.max()
y_min, y_max = x.b.min(), x.b.max()
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))

cm = plt.cm.RdBu
y_tmp = [1 - x for x in y]
```

```python
def best_params_svm(x, y, num_folds=10, score='roc_auc'):
    Cs = [0.01, 0.1, 1, 10, 20, 30, 50, 100]
    gammas = [0.001, 0.01, 0.1, 0.3, 0.5, 1]
    coef0s = [-2, -1, -0.5, 0, 0.5, 1, 2]
    tols = [0.0001, 0.001, 0.01, 0.1, 0.5]
    shrinkings = [True, False]
    probabilities = [True, False]
    
    param_grid = [{'kernel' : ['rbf'], 'tol': tols, 'C': Cs, 'gamma' : gammas, 'shrinking' : shrinkings, 'probability' : probabilities},
                  {'kernel' : ['linear'], 'tol': tols, 'C': Cs, 'shrinking' : shrinkings, 'probability' : probabilities},
                  {'kernel' : ['poly'], 'tol': tols, 'C': Cs, 'gamma' : gammas, 'coef0' : coef0s, 'shrinking' : shrinkings, 'probability' : probabilities},
                  {'kernel' : ['sigmoid'], 'tol': tols, 'C': Cs, 'gamma' : gammas, 'coef0' : coef0s, 'shrinking' : shrinkings, 'probability' : probabilities}]
                  
                  
    grid_search = GridSearchCV(SVC(kernel='rbf'), param_grid, cv=num_folds, scoring=score)
    grid_search.fit(x, y)
    
    return [grid_search.best_params_, grid_search.best_score_]


params_svm = best_params_svm(x, y)

model_svm = SVC(**(params_svm[0]))
model_svm.fit(x, y)

z_svm = model_svm.decision_function(np.c_[xx.ravel(), yy.ravel()])
z_svm = z_svm.reshape(xx.shape)

plt.contourf(xx, yy, z_svm, cmap=cm, alpha=0.6)

plt.scatter(x.a, x.b, c=y_tmp, cmap=plt.cm.coolwarm)
plt.savefig('svm.png', dpi=600)
plt.show()
```

```python
def best_params_knn(x, y, num_folds=10, score='roc_auc'):
    k_range = list(range(1, 31))
    weight_options = ["uniform", "distance"]
    param_grid = dict(n_neighbors = k_range, weights = weight_options)
    grid_search = GridSearchCV(knn(), param_grid, cv=num_folds, scoring=score)
    grid_search.fit(x, y)
    
    return [grid_search.best_params_, grid_search.best_score_]
    
params_knn = best_params_knn(x, y)


model_knn = knn(**(params_knn[0]))
model_knn.fit(x, y)

z_knn = model_knn.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:, 1]
z_knn = z_knn.reshape(xx.shape)

plt.contourf(xx, yy, z_knn, cmap=cm, alpha=0.6)

plt.scatter(x.a, x.b, c=y_tmp, cmap=plt.cm.coolwarm)
plt.savefig('knn.png', dpi=600)
plt.show()
```

```python
print('AUC')
print('-'*30)
print('SVM_best_score =', params_svm[1])
print('KNN_best_score =', params_knn[1])
```
