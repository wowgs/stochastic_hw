```python
import numpy as np
from numpy import linalg
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('task_1_capital.txt', delimiter='\t')
data = data[['Capital', 'Rental']]
```

```python
ar = []
for x in data.Capital:
    ar.extend([1, x])
ar = np.array(ar)
ar.shape = (len(ar) // 2, 2)

y = np.array(data.Rental)
```

```python
dot = np.dot
inv = linalg.inv

w = dot(dot(inv(dot(ar.T, ar)), ar.T), y)
my_y = dot(ar, w)
```

```python
l = len(my_y)
for i in range(len(data)):
    plt.scatter(data.Capital[i], data.Rental[i], c='blue', alpha=0.5, s=20)
plt.plot([data.Capital[0], data.Capital[l-1]], [my_y[0], my_y[l - 1]], c='black')
plt.savefig('res.png', dpi=600)
plt.show()
```
